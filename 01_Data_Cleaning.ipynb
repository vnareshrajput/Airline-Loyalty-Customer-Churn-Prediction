```python
def detect_outliers_iqr(df, col):

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[
        (df[col] < lower) |
        (df[col] > upper)
    ]

    return outliers, Q1, Q3, lower, upper
```

```python
detect_outliers_iqr(loyalty, "Salary")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

outliers, Q1, Q3, lower, upper = detect_outliers_iqr(loyalty, "Salary")

plt.figure(figsize=(12,8))

# Full distribution
sns.histplot(loyalty["Salary"], kde=True, bins=30)

# Highlight middle 50% (Q1 to Q3)
plt.axvspan(Q1, Q3, alpha=0.3, label='25%-75% Range (IQR)')

# Draw percentile lines
plt.axvline(Q1, linestyle='--', linewidth=2, label='Q1 (25%)')
plt.axvline(Q3, linestyle='--', linewidth=2, label='Q3 (75%)')

plt.title("Salary Distribution with IQR Highlighted")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.legend()

plt.show()
```
